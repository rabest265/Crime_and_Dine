```python
import os
import csv
import pandas as pd
```

```python
csvpath = os.path.join("Resources", "crime_data_w_population_and_crime_rate.csv")
raw_df = pd.read_csv(csvpath, encoding="UTF-8")
raw_df.head(2)
```

```python
list(raw_df)
```

```python
new_df = raw_df.drop(['PART','CPOPARST', 'CPOPCRIM', 'AG_ARRST', 'AG_OFF', 'COVIND',
                       'INDEX', 'MODINDX', 'FIPS_ST', 'FIPS_CTY', 'index', 'EDITION', 'IDNO'], 1)
new_df.head(2)
```

```python
crime_df = new_df.rename(index=str, columns={"county_name": "County", "crime_rate_per_100000": "Crime Rate per 100k",
                                 "MURDER": "Murder", "RAPE": "Rape", "ROBBERY": "Robbery",
                                 "AGASSLT": "Assault", "BURGLRY": "Burglary", "LARCENY": "Larceny", 
                                 "MVTHEFT": "MV Theft", "ARSON": "Arson", "population": "Population"})
crime_df.head(2)
```

```python
crime_df.to_csv("Crime_data.csv", index=False, header=True)
```
